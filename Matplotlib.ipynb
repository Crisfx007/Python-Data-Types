```python
import matplotlib.pyplot as plt
```

#### use plt.show() at the end to see plots if not in jupyter but %matplotlib inline shows plots here.

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Functional Method for matplotlib

```python
plt.plot(x,y)
# plt.show() prints the plot
plt.xlabel('Xlabel')
plt.ylabel('ylabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Using object oriented method for plots.

```python
fig = plt.figure()
fig.patch.set_facecolor('lightblue')
axes = fig.add_axes([0.2,0.2,0.8,0.8])
axes.set_facecolor('lightyellow')
axes.plot(x,y)
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger')
axes2.plot(y,x)
axes2.set_title('Smaller')
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('firstplot')
axes[1].set_title('secondplot')
plt.tight_layout()
```
